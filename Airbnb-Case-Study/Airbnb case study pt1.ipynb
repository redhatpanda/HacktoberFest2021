```python
#task 13 what is satisfaction ratings count where price is greater than equal to 150 but less than equal to 180
new_df = airbnb[(airbnb['price'] >= 150) & (airbnb['price'] <=180 )] 
count = new_df['overall_satisfaction'].count()
```

```python
#task 14 Create a copy of main airbnb dataset and impute the value 4 where ever the satisfaction is count 0
copy_dataset = airbnb[airbnb.overall_satisfaction == 0]
```

```python
copy_dataset['overall_satisfaction']= 4
```

```python
copy_dataset.overall_satisfaction.count()
```

```python
#Task 15 Now take the count of satisfaction ratings again after this imputation and compare with main dataset where still 0 is present
print(f"Satisfaction count from the original data: {airbnb.overall_satisfaction.count()}")
print(f"Satisfaction count from the sub dataset {copy_dataset.overall_satisfaction.count()}")
```
